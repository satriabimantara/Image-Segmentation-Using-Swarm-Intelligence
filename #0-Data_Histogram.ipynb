```python
import pickle
import cv2
import matplotlib.pyplot as plt
```

```python
test_dataset_misc = pickle.load(open('datasets/test_dataset.pkl', 'rb'))
gray_misc_dataset = test_dataset_misc['gray']
rgb_misc_dataset = test_dataset_misc['rgb']
```

```python
gray_misc_dataset.keys()
```

```python
for image_name, image_array in gray_misc_dataset.items():
    gray_hist = cv2.calcHist([image_array], [0], None, [256], (0, 256))  # the method can do multiple images at a time

    # Plot of the grayscale image hist
    plt.plot(gray_hist, c='k')
    #plt.title('Grayscale {} intensity hist'.format(image_name))
    plt.xlabel('Pixel Intensity value', fontsize=15)
    plt.ylabel('Number of Pixels', fontsize=15)
    plt.show();
    
```

```python
for image_name, image_array in rgb_misc_dataset.items():
    # Hists of the color image, each channel, same args
    R_hist = cv2.calcHist([image_array], [0], None, [256], (0, 256))
    G_hist = cv2.calcHist([image_array], [1], None, [256], (0, 256))
    B_hist = cv2.calcHist([image_array], [2], None, [256], (0, 256))
    
    # Plot of the RGB image hists
    plt.plot(R_hist, c='r', label='R')
    plt.plot(G_hist, c='g', label='G')
    plt.plot(B_hist, c='b', label='B')
    
    
    # Plot of the grayscale image hist
    plt.xlabel('Pixel Intensity value', fontsize=15)
    plt.ylabel('Number of Pixels', fontsize=15)
    plt.legend()
    plt.show();
```
